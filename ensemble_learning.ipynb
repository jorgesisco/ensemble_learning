# Ensemble Learning
Load the `mnist` dataset. Split it into training and test sets. Train and test an ensemble method, such as random forest or a gradient boosting model, using scikit-learn. Check the documentation to identify the most important hyperparameters, attributes, and methods of the model. Use them in practice.

```python
import pandas as pd
import sklearn.metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import plotly.express as px
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
import numpy as np
import time
```

```python
df = pd.read_csv('/mnist.csv')

# df = df.set_index('id')

print(df.shape)
df.head(3) 
```

```python
x = df.drop(['class'], axis=1)

y = df['class']

x_train, x_test, y_train, y_test, = train_test_split(x,y)

print('df', df.shape)
print('x_train', x_train.shape)
print('x_test', x_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

# Model Selection and Hyperparameter Tunning

```python
# Random Forest ----------------------------------------------------------------
tl_1= time.time()
parameters_grid = {
    'criterion': ['gini', 'entropy'],
    'n_estimators': range(50, 260, 50), # [50, 100, 150, 200, 250]
}

model_1 = GridSearchCV(RandomForestClassifier(),
                 parameters_grid, scoring='accuracy' ,verbose =1, cv=5, n_jobs=-1)

model_1.fit(x_train, y_train)
print('Accuracy of the best Random Forest Classifier: {:.2f}%'.format(model_1.best_score_*100))
print("Best found hyperparameter of Random Forest Classifier: {}".format(model_1.best_params_))

tl_1_done = time.time()
print('Time to get Score and best model params with random forest: {:.2f} seconds'.format( tl_1_done - tl_1))
# -----------------------------------

# Gradient Boosting -----------------------------------
tl_2= time.time()
parameters_grid = {
    'learning_rate': [0.01, 0.05, 0.1, 0.15, 0.20],
    'n_estimators': np.arange(50, 260, 50)
}
model_2 = GridSearchCV(GradientBoostingClassifier(),
                 parameters_grid, scoring='accuracy' , verbose =1, cv=5, n_jobs=-1)

model_2.fit(x_train, y_train)

print('Accuracy of the Gradient Boosting Classifier: {:.2f}%'.format(model_2.best_score_*100))
print("Best found hyperparameter of Gradient Boosting Classifier: {}".format(model_2.best_params_))


tl_2_done = time.time()

print('Time to get Score and best model params with Gradient Boosting Classifier: {:.2f} seconds'.format(tl_2_done - tl_2))
```
